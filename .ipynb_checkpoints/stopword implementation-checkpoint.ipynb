```python
#import required packages
import nltk
import pandas as pd
import csv
import re
import numpy as np
import os
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import ngrams
from nltk.stem import WordNetLemmatizer 
```

```python
#Add more stopwords
adtl_stopwords = pd.read_csv("./preprocessing/test_stopwords.csv", header = None)
adtl_stopwords = adtl_stopwords.values.tolist()
adtl_stopwords = adtl_stopwords[0]
```

```python
stopwords_def = stopwords.words('english')
stopwords_def.extend(adtl_stopwords)
print(stopwords_def)
```

```python
content = open(r"./data/TEST1.txt", "r")
content = content.read()
```

```python
### Remove punctuation and numbers
content = re.sub("[0-9]", "", content) #Removes numbers
content = re.sub(r"\/", " ", content) #Replaces slashes with spaces
content = re.sub(r"[.,?!()-;:@§%&~'\[\]\"]", "", content) #Removes extraneous punctuation
content = re.sub(r"\n","", content) #Removes extraneous line breaks
content = re.sub(r"^$n','\n", "", content, re.MULTILINE) #Removes empty line
content = re.sub(" +", " ", content) #Removes multiple whitespaces
content = re.sub(r"\b\w{1,2}\b", '', content) #Removes all one and two-character words (none have meaning)
content = re.sub(r"([a-z])([A-Z])", r"\1 \2", content) #Removes words that were joined together by double spaces

#Remove extraneous information about lawyers and case detail that doesn't impact the content of the case

starting_phrase = "PRELIMINARY STATEMENT" #Use this as the starter for when the actual lawsuit begins
remove_before = r'^.*?{}'.format(re.escape(starting_phrase))
content = re.sub(remove_before, "", content, flags=re.DOTALL)

#Remove "FIRST AMENDED COMPLAINT", which is at the bottom of every page
content = re.sub(r"FIRST AMENDED COMPLAINT", "", content)
print(content[0:3000])
```

```python
#Word tokenization
#Move to lower
content = content.lower()
word_content = word_tokenize(content)
tokenized_draft = [words for words in word_content if not words.lower() in stopwords_def]
```

```python
#Lemmatization
lemmatizer = WordNetLemmatizer()

lemmatized_output = [lemmatizer.lemmatize(word) for word in tokenized_draft]

print(lemmatized_output)
```
